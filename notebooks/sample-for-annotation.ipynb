```python
import random
import pandas as pd
from textentlib.utils import load_or_create_corpus, nlp_model_fr
```

## First round of sampling

```python
SPACY_CORPUS_SERIALIZED_PATH = "../data/corpus_EF.spacy"
SAMPLE_SIZE = 60
```

```python
spacy_corpus = load_or_create_corpus(SPACY_CORPUS_SERIALIZED_PATH)
```

```python
docs = spacy_corpus.get_docs(nlp_model_fr.vocab)
```

```python
docs = list(docs)
```

```python
sampled_docs = random.sample(docs, SAMPLE_SIZE)
```

```python
sampled_docs[0].user_data
```

```python
df = pd.DataFrame([doc.user_data for doc in sampled_docs]).set_index("document_id", drop=False)
```

```python
df_2 = pd.DataFrame([
    {
        "document_id": doc.user_data["document_id"],
        "document_length": len(doc.text),
    }
    for doc in sampled_docs
]).set_index("document_id")
```

```python
df_2
```

```python
df
```

```python
sample_df = pd.concat([df, df_2], axis=1)
```

```python
sample_df.document_length.describe()
```

```python
sample_df.drop(columns=["path", "filename", "entity_linking"], inplace=True)
```

```python
sample_df
```

```python
sample_df['link_OCR'] = sample_df.document_id.apply(lambda x: f"https://github.com/TextEnt/chrono-spatial-processing/blob/main/data/corpus_normalized/{x}.txt")
```

```python
sample_df.to_csv("../data/sample_60_docs.csv", index=False)
```

## Second round of sampling

Some documents in the first sample had to be excluded (either because they were anthologies or because they were duplicates).
Constraints:
- exclude docs already sampled
- exclude docs w/ > 150k tokens as they are likely to be anthologies

```python
first_sample_df = pd.read_csv("../data/sample_60_docs.csv")
```

```python
first_sample_df.head()
```

```python
doc_ids_first_sample = first_sample_df.document_id.to_list()
```

```python
len(doc_ids_first_sample)
```

```python
SPACY_CORPUS_SERIALIZED_PATH = "../data/corpus_EF.spacy"
SAMPLE_SIZE = 10
```

```python
spacy_corpus = load_or_create_corpus(SPACY_CORPUS_SERIALIZED_PATH)
```

```python
docs = spacy_corpus.get_docs(nlp_model_fr.vocab)
```

```python
docs = list(docs)
```

```python
len(docs)
```

```python
from pathlib import Path
```

```python
non_sampled_docs = [
    doc for doc in docs 
    if (Path(doc.user_data['path']).name.split(".")[0] not in doc_ids_first_sample) and (len(doc.text) <= 100000)
]
```

```python
len(non_sampled_docs)
```

```python
second_sample_docs = random.sample(non_sampled_docs, SAMPLE_SIZE)
```

```python
len(second_sample_docs)
```

```python
for doc in second_sample_docs:
    print(doc.user_data['title'], len(doc.text))
```

```python
second_sample_df = pd.DataFrame([doc.user_data for doc in second_sample_docs])
```

```python
second_sample_df = second_sample_df[['author', 'title', 'publication_date', 'path']]
```

```python
second_sample_df['document_id'] = second_sample_df.path.apply(lambda x: Path(x).name.split(".")[0])
```

```python
second_sample_df.set_index('document_id', inplace=True)
```

```python
df_2 = pd.DataFrame([
    {
        "document_id": Path(doc.user_data['path']).name.split(".")[0],
        "document_length": len(doc.text),
    }
    for doc in second_sample_docs
]).set_index("document_id")
```

```python
sample_df = pd.concat([second_sample_df, df_2], axis=1)
```

```python
sample_df.drop(columns=["path"], inplace=True)
```

```python
sample_df
```

```python
sample_df['link_OCR'] = sample_df.index.map(lambda x: f"https://github.com/TextEnt/chrono-spatial-processing/blob/main/data/corpus_normalized/{x}.txt").to_list()
```

```python
sample_df.to_csv("../data/sample_10_docs.csv")
```
